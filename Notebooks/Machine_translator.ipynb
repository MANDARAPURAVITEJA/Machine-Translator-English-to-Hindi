```python
def english_remover(hindi_text):
    ids_to_remove = {}
    for _id,_t in tqdm(enumerate(hindi_text)):
        if len(re.findall(r'[a-zA-Z]', str(_t))) > 0:
            ids_to_remove[_id] = _t
        else:
            pass
    ids_to_keep = [i for i in range(len(hindi_text)) if i not in ids_to_remove.keys()]
    return ids_to_keep
```
